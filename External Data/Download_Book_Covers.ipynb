### Save Pictures Locally

```python
# load external data
with open('External_Data_incl_Features.pkl', 'rb') as f:
    items_ext = pickle.load(f)
```

```python
items_ext['images'] = pd.DataFrame(items_ext.images.tolist(), index= items_ext.index)
```

```python
list_ext = items_ext.images.value_counts().reset_index()
```

```python
list_ext_df = list_ext.rename( columns={'index' :'link', 'images': 'count'})
```

```python
list_ext_df = list_ext_df.loc[list_ext_df['count']>1]
```

```python
def save_cover_pictures(df, list_check, foldername):
    # df: dataframe where to get links from, 
    # list_check: contains all links that should not be downloaded (preprocessing)
    # foldername: foldername where pictures are saved
    for i in range(52510,len(df)):
        socket.setdefaulttimeout(15)
        print(i)
        image_link = df.iloc[i]['images']
        itemID = df.iloc[i]['itemID']
        if image_link not in list_check:
            try:
                image = image_link
                filename = str(itemID) + '.jpg'
                urllib.request.urlretrieve(str(image), str(foldername) + '/' + str(filename))
                print(str(itemID) + ': link available, pic saved')
            except:
                print('no link available')
        else:
            print(str(itemID)+ ': link in list, do not save picture')
```

```python
# items_ext_wo_duplicates
# links_df_list
folder = 'Cover_Pics_Ext_Final'
list_API = list_ext_df
save_cover_pictures(items_ext, list_API,  folder)
```
